```python
total = train_df.isnull().sum().sort_values(ascending = False)
percent = (train_df.isnull().sum())/(train_df.isnull().count()).sort_values(ascending = False)

missing_data = pd.concat([total, percent], axis = 1, keys = ["total", "percent"])
missing_data
```

```python
train_df = train_df.drop(["keyword", "location", "id"], axis = 1)
print("Keyword, Location, and id are all dropped successfully")

```

```python
##### Turn the text into numbers

freq = {}
for row in train_df["text"]:
    row = row.lower()
    split_row = row.split(" ")
    for word in split_row:
        if word in freq:
            freq[word] += 1
        else:
            freq[word] = 1

print(sorted(freq, key = freq.get, reverse = True))
```
